### Задача 1

Дана строка, состоящая из слов. Сделать из нее аббревиатуру с помощью списковых включений

**Вход:** "Комитет Государственной Безопасности"

**Выход:** "КГБ"

```python
text = "Комитет Государственной Безопасности"
caps = "A<КГБ"
#Решение
b = [i for i in text if i in caps]
print(b)
print(''.join(b))
```
